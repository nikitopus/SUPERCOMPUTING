```python
import matplotlib.pyplot as plt
```

# 1 lab

```python
# ---     ---
vector_sizes = [100_000, 1_000_000, 10_000_000]

time_np4 = [
    0.000194788,  # 100k
    0.00188994,   # 1M
    0.00665903    # 10M
]
time_np8 = [
    0.000118971,
    0.000928879,
    0.00373888
]

#    X   
x_labels = [f"{size:,}".replace(",", " ") for size in vector_sizes]

# ---   ---
plt.figure(figsize=(9, 6))

plt.plot(
    vector_sizes, time_np4,
    label="4 ", marker="o", color="tab:blue",
    linewidth=2, markersize=6
)
plt.plot(
    vector_sizes, time_np8,
    label="8 ", marker="s", color="tab:green",
    linewidth=2, markersize=6
)

# ---  ---
plt.xscale("log")
plt.yscale("log")
plt.xlabel(" ", fontsize=12)
plt.ylabel("  ()", fontsize=12)
plt.title("        ", fontsize=14, fontweight="bold")
plt.xticks(vector_sizes, x_labels, rotation=15)
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12, loc="upper left")

plt.tight_layout()
plt.show()
```

# 2 lab

```python
vector_sizes = [100_000, 1_000_000, 10_000_000, 100_000_000]

time_np4 = [
    0.000113964,  # 100k
    0.00113893,   # 1M
    0.00671005,   # 10M
    0.0707281     # 100M
]

time_np8 = [
    5.6982e-05,
    0.000754118,
    0.00384402,
    0.046313
]

#    X    ( )
x_labels = [f"{size:,}".replace(",", " ") for size in vector_sizes]

# ---   ---
plt.figure(figsize=(9, 6))

plt.plot(
    vector_sizes, time_np4,
    label="4 ", marker="o", color="tab:blue",
    linewidth=2, markersize=6
)
plt.plot(
    vector_sizes, time_np8,
    label="8 ", marker="s", color="tab:green",
    linewidth=2, markersize=6
)

# ---  ---
plt.xscale("log")
plt.yscale("log")
plt.xlabel(" ", fontsize=12)
plt.ylabel("  ()", fontsize=12)
plt.title("        \n  MPI-", fontsize=14, fontweight="bold")
plt.xticks(vector_sizes, x_labels, rotation=15)
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12, loc="upper left")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# ---     ( ) ---
message_sizes_bytes = [
    10,
    100,
    1_000,
    10_000,
    100_000,
    1_000_000,
    10_000_000,
    100_000_000
]

transfer_times_seconds = [
    3.0312e-05,   # 10 B
    7.22e-07,    # 100 B
    2.915e-06,   # 1 KB
    2.7326e-05,  # 10 KB
    2.8378e-05,  # 100 KB
    0.000319476, # 1 MB
    0.00510363,  # 10 MB
    0.0496598    # 100 MB
]

#     X: "10", "100", "1 000", "10 000"  ..
x_labels = [f"{size:,}".replace(",", " ") for size in message_sizes_bytes]

# ---   ---
plt.figure(figsize=(9, 6))

plt.plot(
    message_sizes_bytes, transfer_times_seconds,
    marker="o", color="tab:blue",
    linewidth=2, markersize=6,
    label=" "
)

# ---  ---
plt.xscale("log")
plt.yscale("log")
plt.xlabel("  ()", fontsize=12)
plt.ylabel("  ()", fontsize=12)
plt.title("      \n(2 MPI-)", fontsize=14, fontweight="bold")
plt.xticks(message_sizes_bytes, x_labels, rotation=15)
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
