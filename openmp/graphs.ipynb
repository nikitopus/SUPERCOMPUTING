# 1 lab

```python
import matplotlib.pyplot as plt
import numpy as np

# ---    ---
sizes = [1_000_000, 10_000_000, 100_000_000, 1_000_000_000]
time_with_reduction = [0.00265002, 0.00641608, 0.0689671, 0.458701]
time_without_reduction = [0.000974894, 0.00454497, 0.0496259, 0.459511]

# ---    ---
plt.figure(figsize=(10, 6))

plt.plot(
    sizes, time_with_reduction,
    label=" ", marker="o", linestyle="-", color="tab:blue", linewidth=2, markersize=6
)
plt.plot(
    sizes, time_without_reduction,
    label=" ", marker="s", linestyle="--", color="tab:orange", linewidth=2, markersize=6
)

plt.title("      ", fontsize=14, fontweight="bold")
plt.xlabel(" ", fontsize=12)
plt.ylabel("  ()", fontsize=12)
plt.xscale("log")
plt.yscale("log")
plt.xticks(sizes, [f"{size:,}".replace(",", " ") for size in sizes], rotation=15)
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12, loc="upper left")
plt.tight_layout()
plt.show()

# ---     ---
speedup = np.array(time_with_reduction) / np.array(time_without_reduction)

plt.figure(figsize=(10, 6))
plt.plot(
    sizes, speedup,
    marker="^", linestyle="-", color="tab:green", linewidth=2, markersize=7, label=""
)

plt.title("   ", fontsize=14, fontweight="bold")
plt.xlabel(" ", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xscale("log")
plt.xticks(sizes, [f"{size:,}".replace(",", " ") for size in sizes], rotation=15)
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.7)
plt.axhline(y=1, color="red", linestyle="--", linewidth=1.2, label=" ")
plt.legend(fontsize=12, loc="upper right")
plt.tight_layout()
plt.show()
```
